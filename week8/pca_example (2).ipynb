```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Step 1: Selfie data (Smile Level, Brightness)
# Each row is a selfie with 2 values: [Smile, Brightness]
selfie_data = np.array([
    [2, 3],   # Slight smile, medium brightness
    [3, 5],   # Medium smile, more brightness
    [4, 7],   # Bigger smile, bright!
    [5, 8],   # Even bigger smile, very bright!
    [6, 10]   # Huge smile, super bright!
])

#  Step 2: Plot the original 2D selfie data
plt.scatter(selfie_data[:, 0], selfie_data[:, 1], color='purple')
plt.title('Original Selfie Scores')
plt.xlabel('Smile Level üòÅ')
plt.ylabel('Brightness ‚òÄÔ∏è')
plt.grid(True)
plt.show()

#  Step 3: PCA to reduce from 2D (Smile + Brightness) to 1D (Selfie Mood Score)
pca = PCA(n_components=1)
selfie_mood_score = pca.fit_transform(selfie_data)
print(selfie_mood_score.shape)



# üìâ Step 4: Show the reduced 1D scores
print("\n PCA Results:")
#How much of the total variation in the data is captured by each principal component.
print(f"Explained Variance Ratio: {pca.explained_variance_ratio_}")
print(f"Singular Values: {pca.singular_values_}")
print(pca.components_)
print("Selfie Mood Scores (1D):")
print(selfie_mood_score)

# Step 5: Rebuild the selfies just for plotting
projected_selfies = pca.inverse_transform(selfie_mood_score)

# Step 6: Visualize original and projected data
plt.scatter(selfie_data[:, 0], selfie_data[:, 1], color='purple', label='Original Selfies')
plt.scatter(projected_selfies[:, 0], projected_selfies[:, 1], color='orange', label='PCA Mood Line')
plt.legend()
plt.title('Selfie Mood Line (PCA View)')
plt.xlabel('Smile Level üòÅ')
plt.ylabel('Brightness ‚òÄÔ∏è')
plt.grid(True)
plt.show()
```
