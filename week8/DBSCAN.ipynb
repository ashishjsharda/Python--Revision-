```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Coordinates of kids on a playground
kids = np.array([
    [1, 2], [2, 2], [2, 3],    # Group 1 (close together)
    [8, 7], [8, 8], [25, 80],  # Group 2 and 1 loner far away
])

# Group detection using DBSCAN
model = DBSCAN(eps=2, min_samples=2)
labels = model.fit_predict(kids)
print(labels)


# Plot
plt.scatter(kids[:, 0], kids[:, 1], c=labels, cmap='rainbow')
plt.title("Friend Groups Found by DBSCAN")
plt.xlabel("Playground X")
plt.ylabel("Playground Y")
plt.show()
```
