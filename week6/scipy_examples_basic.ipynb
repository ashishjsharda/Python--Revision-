```python
import numpy as np
from scipy.cluster.hierarchy import linkage, fcluster

# Step 1: Some toy coordinates (like dots on a paper)
data = np.array([[1,2], [1.5,1.8], [5,8], [8,8], [1,0.6], [9,11]])

# Step 2: Group the dots into a tree (hierarchical clustering)
clusters = linkage(data, method='ward')

# Step 3: Cut the tree into 2 groups
labels = fcluster(clusters, 2, criterion='maxclust')

# Step 4: Show which dot belongs to which group
print("Dots:", data)
print("Groups:", labels)
```

```python
import matplotlib.pyplot as plt

# Plot the points and color them by cluster
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='rainbow', s=100)

# Add labels to each point (optional)
for i, point in enumerate(data):
    plt.text(point[0]+0.1, point[1], f"P{i+1}", fontsize=9)

plt.title("Hierarchical Clustering (2 Groups)")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```
