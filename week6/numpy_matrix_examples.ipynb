```python
import numpy as np
import matplotlib.pyplot as plt

print("=== NUMPY MATRICES: FROM BASICS TO REAL WORLD ===\n")

# 1. Creating Matrices
print("1. CREATING MATRICES")
print("=" * 30)

# Different ways to create matrices
matrix_a = np.array([[1, 2, 3],
                     [4, 5, 6]])
print("2x3 Matrix A:")
print(matrix_a)
print(f"Shape: {matrix_a.shape}\n")

# Special matrices
identity = np.eye(3)  # Identity matrix
zeros = np.zeros((2, 3))
ones = np.ones((3, 2))
random_matrix = np.random.randint(1, 10, (3, 3))

print("Identity Matrix (3x3):")
print(identity)
print("\nRandom Matrix (3x3):")
print(random_matrix)

# 2. Matrix Operations
print("\n2. MATRIX OPERATIONS")
print("=" * 30)

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Matrix A:")
print(A)
print("Matrix B:")
print(B)

# Basic operations
print(f"\nA + B (Addition):")
print(A + B)

print(f"\nA * B (Element-wise multiplication):")
print(A * B)

print(f"\nA @ B (Matrix multiplication):")
print(A @ B)

print(f"\nA.T (Transpose of A):")
print(A.T)

# 3. Real-World Example 1: Image Processing
print("\n3. REAL-WORLD EXAMPLE: IMAGE PROCESSING")
print("=" * 50)

# Create a simple 8x8 "image" (grayscale values 0-255)
np.random.seed(42)
image = np.random.randint(0, 256, (8, 8))
print("Original 8x8 Image (grayscale values):")
print(image)

# Apply a blur filter using matrix convolution
blur_kernel = np.ones((3, 3)) / 9  # Average of 3x3 neighborhood
print(f"\nBlur kernel (3x3):")
print(blur_kernel)

# Simple edge detection filter
edge_kernel = np.array([[-1, -1, -1],
                        [-1,  8, -1],
                        [-1, -1, -1]])
print(f"\nEdge detection kernel:")
print(edge_kernel)

# 4. Real-World Example 2: Portfolio Analysis
print("\n4. REAL-WORLD EXAMPLE: INVESTMENT PORTFOLIO")
print("=" * 50)

# Stock returns for 3 stocks over 5 days
stock_returns = np.array([
    [0.02, -0.01, 0.03],  # Day 1: Stock A, B, C returns
    [-0.01, 0.02, 0.01],  # Day 2
    [0.03, -0.005, 0.02], # Day 3
    [0.01, 0.015, -0.01], # Day 4
    [-0.02, 0.01, 0.025]  # Day 5
])

portfolio_weights = np.array([0.4, 0.35, 0.25])  # 40%, 35%, 25%

print("Stock returns matrix (5 days × 3 stocks):")
print(stock_returns)
print(f"\nPortfolio weights: {portfolio_weights}")

# Calculate daily portfolio returns
daily_portfolio_returns = stock_returns @ portfolio_weights
print(f"\nDaily portfolio returns:")
print(daily_portfolio_returns.round(4))

# Calculate correlation matrix
correlation_matrix = np.corrcoef(stock_returns.T)
print(f"\nStock correlation matrix:")
print(correlation_matrix.round(3))

# 5. Real-World Example 3: Machine Learning (Linear Regression)
print("\n5. REAL-WORLD EXAMPLE: MACHINE LEARNING")
print("=" * 50)

# Simple house price prediction based on size and bedrooms
# X matrix: [size_sqft, bedrooms, bias_term]
X = np.array([
    [1500, 3, 1],  # House 1
    [2000, 4, 1],  # House 2
    [1200, 2, 1],  # House 3
    [1800, 3, 1],  # House 4
    [2200, 4, 1]   # House 5
])

# Actual prices (in thousands)
y = np.array([300, 400, 250, 350, 450])

print("House features matrix X (size, bedrooms, bias):")
print(X)
print(f"\nActual prices (thousands): {y}")

# Solve for weights using normal equation: w = (X^T * X)^(-1) * X^T * y
XtX = X.T @ X
XtX_inv = np.linalg.inv(XtX)
Xty = X.T @ y
weights = XtX_inv @ Xty

print(f"\nLearned weights (price per sqft, price per bedroom, base price):")
print(f"${weights[0]:.2f} per sqft, ${weights[1]:.0f}k per bedroom, ${weights[2]:.0f}k base")

# Predict prices
predicted_prices = X @ weights
print(f"\nActual vs Predicted prices:")
for i, (actual, pred) in enumerate(zip(y, predicted_prices)):
    print(f"House {i+1}: ${actual}k actual, ${pred:.0f}k predicted")

# 6. Matrix Properties and Advanced Operations
print("\n6. ADVANCED MATRIX OPERATIONS")
print("=" * 40)

test_matrix = np.array([[4, 2], [1, 3]])
print("Test Matrix:")
print(test_matrix)

# Determinant
det = np.linalg.det(test_matrix)
print(f"\nDeterminant: {det}")

# Eigenvalues and eigenvectors
eigenvals, eigenvecs = np.linalg.eig(test_matrix)
print(f"\nEigenvalues: {eigenvals.round(3)}")
print(f"Eigenvectors:\n{eigenvecs.round(3)}")

# Matrix inverse
inverse = np.linalg.inv(test_matrix)
print(f"\nInverse matrix:")
print(inverse.round(3))

# Verify: A * A^(-1) = Identity
verification = test_matrix @ inverse
print(f"\nVerification (should be identity):")
print(verification.round(10))

# Visualization
plt.figure(figsize=(15, 10))

# Plot 1: Correlation Heatmap
plt.subplot(2, 3, 1)
plt.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar()
plt.title('Stock Correlation Matrix')
plt.xticks([0,1,2], ['Stock A', 'Stock B', 'Stock C'])
plt.yticks([0,1,2], ['Stock A', 'Stock B', 'Stock C'])

# Plot 2: Portfolio Returns
plt.subplot(2, 3, 2)
plt.plot(daily_portfolio_returns, 'bo-', linewidth=2, markersize=8)
plt.title('Daily Portfolio Returns')
plt.xlabel('Day')
plt.ylabel('Return')
plt.grid(True)

# Plot 3: House Price Prediction
plt.subplot(2, 3, 3)
house_numbers = np.arange(1, len(y)+1)
plt.bar(house_numbers - 0.2, y, 0.4, label='Actual', alpha=0.7)
plt.bar(house_numbers + 0.2, predicted_prices, 0.4, label='Predicted', alpha=0.7)
plt.title('House Price Prediction')
plt.xlabel('House')
plt.ylabel('Price ($k)')
plt.legend()

# Plot 4: Original "Image"
plt.subplot(2, 3, 4)
plt.imshow(image, cmap='gray')
plt.title('Original 8x8 "Image"')
plt.colorbar()

# Plot 5: Eigenvalue visualization
plt.subplot(2, 3, 5)
# Plot eigenvectors as arrows
plt.quiver(0, 0, eigenvecs[0,0], eigenvecs[1,0], angles='xy', scale_units='xy', scale=1, color='red', width=0.005)
plt.quiver(0, 0, eigenvecs[0,1], eigenvecs[1,1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.005)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.grid(True)
plt.title('Eigenvectors')
plt.xlabel('X')
plt.ylabel('Y')

# Plot 6: Matrix transformation visualization
plt.subplot(2, 3, 6)
# Original unit square
square = np.array([[0, 1, 1, 0, 0], [0, 0, 1, 1, 0]])
# Transform the square
transformed = test_matrix @ square
plt.plot(square[0], square[1], 'b-', linewidth=2, label='Original')
plt.plot(transformed[0], transformed[1], 'r-', linewidth=2, label='Transformed')
plt.title('Matrix Transformation')
plt.legend()
plt.grid(True)
plt.axis('equal')

plt.tight_layout()
plt.show()

print("\n=== MATRIX APPLICATIONS SUMMARY ===")
print("• Image Processing: Filters and convolutions")
print("• Finance: Portfolio optimization and risk analysis")
print("• Machine Learning: Linear regression, neural networks")
print("• Computer Graphics: 3D transformations and rotations")
print("• Engineering: Solving systems of equations")
print("• Data Science: Principal Component Analysis (PCA)")
```
