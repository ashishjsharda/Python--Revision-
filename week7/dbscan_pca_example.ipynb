```python
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# 1. Generate sample data (500 samples, 10 features, 3 centers)
X, _ = make_blobs(n_samples=500, n_features=10, centers=3, random_state=42)

# 2. Apply PCA to reduce dimensions to 2
X_pca = PCA(n_components=2).fit_transform(X)

# 3. Apply DBSCAN on the PCA-reduced data
dbscan = DBSCAN(eps=0.5, min_samples=5).fit(X_pca)

# 4. Plot the results (simplified, assuming 2D for visualization)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=dbscan.labels_)
plt.title("DBSCAN on PCA-reduced Data")
plt.show()
```
