```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 1. Load sample high-dimensional data (Iris dataset has 4 features)
# load_iris() is a scikit-learn utility function that loads the famous Iris flower dataset.
# This dataset contains measurements of 150 Iris flowers, each with 4 features:
# sepal length, sepal width, petal length, and petal width.
# It also includes the species (target) for each flower (setosa, versicolor, virginica).
data = load_iris()

# data.data accesses the feature measurements (the 4 numerical features for each flower).
# X will be a NumPy array of shape (150, 4) – 150 samples, 4 features.
X = data.data

# 2. Apply PCA to reduce dimensions to 2
# Initialize the PCA model:
# PCA is a dimensionality reduction technique that transforms a set of correlated variables
# into a set of linearly uncorrelated variables called principal components.
# n_components=2 tells PCA to reduce the data down to only 2 principal components.
# These 2 components will capture the most variance (i.e., information) from the original 4 features.
pca = PCA(n_components=2)

# Perform PCA:
# fit_transform(X) first 'fits' the PCA model to the data X. This means it calculates
# the principal components based on the variance in X.
# Then, it 'transforms' the original data X into the new lower-dimensional space defined by these components.
# X_pca will be a NumPy array of shape (150, 2) – 150 samples, now with only 2 features (principal components).
X_pca = pca.fit_transform(X)

# 3. Plot the PCA-transformed data to visualize
# Create a 2D scatter plot using Matplotlib.
# plt.scatter(): This function draws individual data points as markers on a 2D plane.
# X_pca[:, 0]: Uses the values of the first principal component for each data point as the x-axis values.
# X_pca[:, 1]: Uses the values of the second principal component for each data point as the y-axis values.
# c=data.target: This is key! It colors each data point based on its original flower species (its true label).
#                Since the Iris dataset has 3 species, you'll see 3 different colors on the plot.
#                This allows you to visually check if PCA effectively separated the different species in 2D.
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data.target)
plt.title("Iris Data Reduced with PCA")
plt.xlabel("Principal Component 1") # Label the x-axis to reflect the transformed data.
plt.ylabel("Principal Component 2") # Label the y-axis to reflect the transformed data.
plt.show() # Display the generated plot.
```
