```python
from skimage import data, filters
import matplotlib.pyplot as plt

image = data.coins() # Loads a sample image (coins)
edges = filters.sobel(image)


plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import requests
from IPython.display import Image, display

# Step 1: Search NASA Image API for "moon"
url = "https://images-api.nasa.gov/search?q=moon&media_type=image"
response = requests.get(url)
data = response.json()
print(data)

# Step 2: Get the first image's URL
items = data['collection']['items']
if items:
    first_item = items[0]
    image_url = first_item['links'][0]['href']
    print("Image URL:", image_url)

    # Step 3: Display in Colab
    display(Image(url=image_url))
else:
    print("No moon images found.")

```
